```python
import pandas as pd
import numpy as np

#load the excel file
loc_tmp = ("E:\\Energy Indicators.xls")
file=pd.read_excel(loc_tmp)

#create dataframe
energy=pd.DataFrame(file)

#remove header and footer
energy=energy.drop(energy.index[0:17])
energy=energy.drop(energy.index[227:])
energy.reset_index(drop=True,inplace=True)

#rename columns
energy.columns=['tmp1','tmp2','Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']

#drop the first two columns
energy.drop('tmp1',axis=1,inplace=True)
energy.drop('tmp2',axis=1,inplace=True)

energy['Energy Supply']=energy['Energy Supply']*1000000

# For all countries which have missing data (e.g. data with "...") make sure this is reflected as np.NaN values.

energy_row_count=energy.shape[0]
i=0

while(i < energy_row_count):
    
    #set Nan in Energy Supply Column
    if ("str" in str(type(energy['Energy Supply'][i]))):
        energy['Energy Supply'][i]=np.nan
    
    #set Nan in Energy Supply per Capita Column
    if ("str" in str(type(energy['Energy Supply per Capita'][i]))):
        energy['Energy Supply per Capita'][i]=np.nan

    i+=1  

energy

if(energy['Country']=="Republic of Korea"):
    energy['Country']
    

: "South Korea",
"United States of America": "United States",
"United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
"China, Hong Kong Special Administrative Region": "Hong Kong"
    
    
```
