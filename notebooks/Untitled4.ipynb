```python
import re
str="Four score <body and seven years ago our fathers brought forth on this continent, a new nation, conceived in Liberty, and dedicated to the proposition that all men are created equal. Now </body>"
x=re.split("<body",str)[1]
y=re.split("</body>",x)[0]
for j in y.split():
    print("j is",j)
    t=re.sub("\.","",j)
    print("t is",t)
    i=re.sub("\,","",t)
    print("i is",i)
```
