```python
import re
import urllib.request
from bs4 import BeautifulSoup
 
#html = urllib.request.urlopen('https://www.vox.com/2016/7/21/12253426/donald-trump-acceptance-speech-transcript-republican-nomination-transcript')
html = urllib.request.urlopen('https://lifefastlaneblog.wordpress.com/')
#html = urllib.request.urlopen('https://101balance.wordpress.com/')

soup = BeautifulSoup(html)
data = soup.findAll(text=True)
 
def visible(element):
    if element.parent.name in ['style', 'script', '[document]', 'head', 'title']:
        return False
    elif re.match('<!--.*-->\\n', str(element.encode('utf-8'))):
        return False
    return True
 
result = filter(visible, data)

str1=""
for text in result:
    str1+=text
#print(str1)

#print(result)
 
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 

filt={"be","you","He","his", "those", "an","she","at","all","but","what","\n","the","with","her","was","and","this","for","of","to","on","not","that","as","it","in","is","Barack Obama","Barack","Obama","who","there","are","have","by","has", "from","which"}

wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = filt, 
                min_font_size = 10).generate(str1)

plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```
